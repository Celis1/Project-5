```python
import pandas as pd
```

```python
tags = pd.read_csv('./Assets/TAGS_disastertwitter_data')
```

```python
tags.head()
```

```python
import nltk
import regex as re
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import RegexpTokenizer
```

```python
tags.head(3)
```

```python
#making a list of tweets
tweets = []
for i in tags.text:
    tweets.append(i)
tweets
```

```python
#definig tokenizer and lemmatizer
tokenizer = RegexpTokenizer(r'\w+')
lemmatizer = WordNetLemmatizer()
```

```python
#tokenizes and lemmatizes the words in each tweet
tokens = []
tweet_dict = {}
count = 0
for i in tweets:
    meh = tokenizer.tokenize(i.lower())
    lem = [lemmatizer.lemmatize(i) for i in meh]
    tweet_dict[count] = lem
    count += 1
```

## fema function

```python
df = pd.read_csv('./fema.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
for i in ['Headlines','Disaster Type','Photo Date','Report Date']:
    df[i] = df[i].apply(lambda x : x.lower())
```

```python
df.head()
```

```python
disaster_words = ['road','neighbor','street','hwy','highway']
```

```python
#functions for extracting info from df
def dict_of_keywords(df,search_words):
    info = {}
    headline = {}
    for i in range(len(df)):
        for n in search_words:
            if n in df['Headlines'][i]:
                headline.update({'Headlines':df.iloc[i]['Headlines']})
                headline.update({'Disaster_Type':df.iloc[i]['Disaster Type']})
                headline.update({'State':df.iloc[i]['State']})
                headline.update({'Report_data':df.iloc[i]['Report Date']})
                info[i] = headline
                headline = {}
    news_df = pd.DataFrame(info,columns=info.keys()).T
    return(news_df)
```
